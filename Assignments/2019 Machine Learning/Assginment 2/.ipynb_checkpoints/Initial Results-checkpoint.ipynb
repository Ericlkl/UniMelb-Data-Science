```python
# Import libraries
import pandas as pd
import numpy as np

# Import metrics and evaluations
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.svm import LinearSVC # one vs rest
from sklearn.linear_model import LogisticRegression 
# solver='lbfgs', multi_class='multinomial', max_iter=1000
```

## Helper Functions

```python
category = ['Sydney', 'Melbourne', 'Brisbane', 'Perth']
datasets = ['10', '50', '100']

def report(clf, X_test, y_test):
    y_pred = clf.predict(X_test)
    print(classification_report(y_test, y_pred,target_names=category))
    print(f'Accuracy: {100*accuracy_score(y_pred, y_test):.2f}%')
    df = pd.DataFrame(confusion_matrix(y_test, y_pred, labels=category), index=category, columns=category)
    sns.heatmap(df, annot=True, fmt='d')
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.yticks(rotation=0)
    plt.show()
    return y_pred

# Returns train / test datasets with respect to its top-n dataset.
def preprocess(top='10'):
    if top == '10':
        data = pd.read_csv("train-top10.csv", index_col = 0)
        test_data = pd.read_csv("dev-top10.csv", index_col=0)
    elif top == '50':
        data = pd.read_csv("train-top50.csv", index_col = 0)
        test_data = pd.read_csv("dev-top50.csv", index_col=0)
    elif top == '100':
        data = pd.read_csv("train-top100.csv", index_col = 0)
        test_data = pd.read_csv("dev-top100.csv", index_col=0)
    else:
        return None
    X_train, y_train = data[data.columns[:-1]], data['Location']
    X_test, y_test = test_data[data.columns[:-1]], test_data['Location']
    return X_train, y_train, X_test, y_test
```

```python
y_pred = pd.DataFrame()
```

## Naive Bayes
- Multinomial

```python
for i in datasets:
    print(f"Training using Top-{i} data set")
    X_train, y_train, X_test, y_test = preprocess(i)
    # clf = MultinomialNB()
    # best_params = random_search_tune(X_train, y_train, clf, param)
    # alpha = list(best_params.values())[0]
    clf = MultinomialNB().fit(X_train, y_train)
    y_pred['nb'+i] = report(clf, X_test, y_test)
```

## SVM
- Using a one vs rest

```python
for i in datasets:
    print(f"Training using Top-{i} data set")
    X_train, y_train, X_test, y_test = preprocess(i)
    clf = LinearSVC(max_iter=10000).fit(X_train, y_train)
    y_pred['svm'+i] = report(clf, X_test, y_test)
```

## Logistic Regression


```python
for i in datasets:
    print(f"Training using Top-{i} data set")
    X_train, y_train, X_test, y_test = preprocess(i)
    clf = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=1000).fit(X_train, y_train)
    y_pred['lr'+i] = report(clf, X_test, y_test)
```

## Stacked Ensemble Learner
- `clf1` = MultinomialNB (X_train_freq, X_dev_freq, X_test_freq)
- `clf2` = LinearSVC (X_train_tfidf, X_dev_tfidf, X_test_tfidf)
- `clf3` = LogisticRegression (X_train_freq, X_dev_freq, X_test_freq)

```python
from mlxtend.classifier import StackingCVClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
stacked_clf1 = RandomForestClassifier(n_estimators=100, min_samples_split=50, 
                              min_samples_leaf=1, criterion='gini', n_jobs=-1)
stacked_clf2 = MultinomialNB()
stacked_clf3 = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=1000, n_jobs=-1)
lr = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=1000, n_jobs=-1)

sclf = StackingCVClassifier(classifiers=[stacked_clf1, stacked_clf2, stacked_clf3],
                            meta_classifier=lr, use_probas=True)

for i in datasets:
    X_train, y_train, X_test, y_test = preprocess(i)
    
    sclf.fit(X_train, y_train.replace(category,[0,1,2,3]))

    preds = pd.Series(sclf.predict(X_test)).replace([0,1,2,3], category)
    print(classification_report(y_test, preds,target_names=category))
    print(f'Accuracy: {100*accuracy_score(preds, y_test):.2f}%')
    df = pd.DataFrame(confusion_matrix(y_test, preds, labels=category), index=category, columns=category)
    sns.heatmap(df, annot=True, fmt='d')
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.yticks(rotation=0)
    plt.show()
```
