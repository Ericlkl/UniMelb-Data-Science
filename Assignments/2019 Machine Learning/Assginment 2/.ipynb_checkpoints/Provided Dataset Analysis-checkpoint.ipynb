```python
# Import libraries
import pandas as pd
import numpy as np

# Import metrics and evaluations
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Hyper Parameter Tuning
from sklearn.model_selection import RandomizedSearchCV

# Import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.svm import LinearSVC # one vs rest
from sklearn.linear_model import LogisticRegression 
# solver='lbfgs', multi_class='multinomial', max_iter=1000
```

## Helper Functions

```python
category = ['Sydney', 'Melbourne', 'Brisbane', 'Perth']
datasets = ['10', '50', '100']

def report(clf, X_test, y_test):
    y_pred = clf.predict(X_test)
    print(classification_report(y_test, y_pred,target_names=category))
    print(f'Accuracy: {100*accuracy_score(y_pred, y_test):.2f}%')
    df = pd.DataFrame(confusion_matrix(y_test, y_pred, labels=category), index=category, columns=category)
    sns.heatmap(df, annot=True, fmt='d')
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.yticks(rotation=0)
    plt.show()

# Returns train / test datasets with respect to its top-n dataset.
def preprocess(top='10'):
    if top == '10':
        data = pd.read_csv("train-top10.csv", index_col = 0)
        test_data = pd.read_csv("dev-top10.csv", index_col=0)
    elif top == '50':
        data = pd.read_csv("train-top50.csv", index_col = 0)
        test_data = pd.read_csv("dev-top50.csv", index_col=0)
    elif top == '100':
        data = pd.read_csv("train-top100.csv", index_col = 0)
        test_data = pd.read_csv("dev-top100.csv", index_col=0)
    else:
        return None
    X_train, y_train = data[data.columns[:-1]], data['Location']
    X_test, y_test = test_data[data.columns[:-1]], test_data['Location']
    return X_train, y_train, X_test, y_test

# Function that uses Randomized Serach to tune Hyper-Parameters.
def random_search_tune(X_train, y_train, estimator, parameters):
    rs = RandomizedSearchCV(estimator, parameters, n_iter=50, cv=3, random_state=42, n_jobs=-1)
    rs_fit = rs.fit(X_train, y_train)
    display(rs_fit.best_params_)
    return rs_fit.best_params_
```

## Naive Bayes
- Multinomial

```python
# param = {'alpha': [i/100 for i in range(1, 100)]}
alphas = {'10': 0.63, '50': 0.41, "100": 0.11}

for i in datasets:
    print(f"Training using Top-{i} data set")
    X_train, y_train, X_test, y_test = preprocess(i)
    # clf = MultinomialNB()
    # best_params = random_search_tune(X_train, y_train, clf, param)
    # alpha = list(best_params.values())[0]
    clf = MultinomialNB(alphas[i]).fit(X_train, y_train)
    report(clf, X_test, y_test)
```

## SVM
- Using a one vs rest

```python
for i in datasets:
    print(f"Training using Top-{i} data set")
    X_train, y_train, X_test, y_test = preprocess(i)
    clf = LinearSVC(max_iter=10000).fit(X_train, y_train)
    report(clf, X_test, y_test)
```

## Logistic Regression


```python
for i in datasets:
    print(f"Training using Top-{i} data set")
    X_train, y_train, X_test, y_test = preprocess(i)
    clf = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=1000).fit(X_train, y_train)
    report(clf, X_test, y_test)
```
