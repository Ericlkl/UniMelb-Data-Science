```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
ef = pd.ExcelFile('time.xlsx')
```

```python
ef.sheet_names
```

```python
df1 = ef.parse('Non Optimised').set_index('Puzzle (N)')
df2 = ef.parse('Optimised').set_index('Puzzle (N)')
df3 = ef.parse('WIDA + Optimised').set_index('Puzzle (N)')
```

```python
plt.rcParams['figure.figsize'] = (15, 10)
sns.set(font_scale=1.2)
for i in df1.columns:
    fig = plt.figure()
    ax=pd.concat([df1[i],df2[i],df3[i]],axis=1).plot.barh()
    plt.title(str(i))
    plt.legend(ef.sheet_names)
    plt.savefig(str(i.split('/'))+'.png')
```
